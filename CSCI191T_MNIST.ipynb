# MLP Classifier

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix,classification_report
```

```python
# Load training and validation sets
df_train = pd.read_csv('./data/train.csv')
df_test = pd.read_csv('./data/test.csv')
```

```python
df_train.head()
```

I want to split the images and labels as X and y

```python
X_train = df_train.drop('label', axis=1)
y_train = df_train['label']

X_test = df_test
```

```python
#Convert from Pandas DataFrame to Numpy Array to be able to perform reshape operations in the next step
X_train = X_train.to_numpy()
y_train = y_train.to_numpy()

X_test = X_test.to_numpy()
```

```python
X_train = X_train.reshape((-1,28,28))
X_test = X_test.reshape((-1,28,28))
```

Visualize the pixel data

```python
import seaborn as sns
sns.countplot(x=y_train)
```

```python
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.imshow(X_train[i])
plt.show()
```

```python
X_train = X_train.reshape((-1,28*28))
X_test = X_test.reshape((-1,28*28))
```

Our initial model

I am going to use hidden layer sizes of 64x64

set early_stopping to true so that we don't overtrain 

set verbose to true because we want log statements

```python
model = MLPClassifier(solver='adam', activation='relu', hidden_layer_sizes = (64,64), early_stopping=True, verbose=True)
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
plt.figure(figsize=(15,6))
for i in range(40):  
    plt.subplot(4, 10, i+1)
    plt.imshow(X_test[i].reshape((28,28)),cmap=plt.cm.binary)
    plt.title("predict=%d" % pred[i],y=0.9)
    plt.xticks([])
    plt.yticks([])
plt.subplots_adjust(wspace=0.3, hspace=-0.1)
plt.show()
```
